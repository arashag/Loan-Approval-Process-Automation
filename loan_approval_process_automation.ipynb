From above 6 cells we see that the only tables that we need to use aggregation function on their variables are transaction and orders. But because we have a lot of missing values in the order table for around 800 account IDs, the order table is not included in our analysis.The rest tables can be merged after grouping and aggregation of these tables

```python
#This function calculates how long ago an account is opened
def duration(x):
    return datetime.strptime("1998-12-31",'%Y-%m-%d')-x.min()
```

```python
#Focusing on those accounts that receive loans
fin_loan_df=fin_df[~fin_df['loan_id'].isna()]
fin_loan_df.head()
```

```python
fin_final_df=fin_loan_df.groupby('account_id').agg({'acc_date':duration,'district_id':'min','trans_amount':'mean','balance':'mean'})
fin_final_df['acc_date']=fin_final_df['acc_date'].dt.days
fin_final_df.columns=['account_age','district_id','mean_trans_amount','mean_balance']
fin_final_df.reset_index(inplace=True)
fin_final_df.head()
```

```python
fin_final_df.shape
```

```python
#Total count of each category
trans_agg_df=fin_loan_df.groupby('account_id')['trans_type'].value_counts().unstack().fillna(0)
trans_agg_df=trans_agg_df.merge(fin_loan_df.groupby('account_id')['operation'].value_counts().unstack().fillna(0),left_index=True,right_index=True)
trans_agg_df=trans_agg_df.merge(fin_loan_df.groupby('account_id')['trans_k_symbol'].value_counts().unstack().fillna(0),left_index=True,right_index=True)
```

```python
#Based on what I find on the internet, some column names are translated
trans_agg_df.drop(columns=[' ','VYBER_x'],inplace=True)
trans_agg_df.columns=['trans_credit','trans_debit','transfer_to','transfer_from','credit_in_cash','cash_withdrawal',
                      'credit_card_withdrawal','insurance_payment','sanction_interest','household_payment','statement_payment',
                     'credited_interest','loan_payment']
trans_agg_df.reset_index(inplace=True)
```

```python
trans_agg_df.head()
```

```python
disp_client_df=disp_df[disp_df['type']=='OWNER'].merge(client_df,on='client_id')
```

```python
disp_client_df['birth_date']=pd.to_datetime(disp_client_df['birth_date'],format='%Y-%m-%d')
```

```python
#Calculating Age of each account owner as of 1998-12-31
disp_client_df['client_age']=(datetime.strptime("1998-12-31",'%Y-%m-%d')-disp_client_df['birth_date']).dt.days/365
```

```python
#removing some non useful columns
disp_client_df.drop(columns=['disp_id','client_id','type','birth_date','district_id'],inplace=True)
```

```python
disp_client_df.head()
```

```python
loan_edited_df=loan_df.copy()
loan_edited_df.set_index('account_id',inplace=True)
loan_edited_df['account_age_at_loan']=(fin_loan_df.groupby('account_id')['acc_date','loan_date'].min()['loan_date']-
                                       fin_loan_df.groupby('account_id')['acc_date','loan_date'].min()['acc_date']).dt.days
loan_edited_df.drop(columns=['loan_id','date'],inplace=True)
loan_edited_df.reset_index(inplace=True)
```

```python
loan_edited_df.head()
```

```python
#Merging all the processed tables
fin_final_df=fin_final_df.merge(disp_client_df,on='account_id')
fin_final_df=fin_final_df.merge(dist_df,on='district_id')
fin_final_df=fin_final_df.merge(loan_edited_df,on='account_id')
fin_final_df=fin_final_df.merge(trans_agg_df,on='account_id')
```

```python
fin_final_df.head()
```

## Analysis and Modelling

```python
fin_final_df=pd.read_csv('./fin_final_df.csv'),header=0)
```

```python
# A & C status is accepted and B & D are rejected
def target_creator(x):
    if x=='A' or x=='C':
        return 1
    if x=='B' or x=='D':
        return 0
```

```python
y=fin_final_df['status'].apply(target_creator)
y.head()
```

```python
y.value_counts()
```

```python
X=fin_final_df.drop(columns=['account_id','district_id','status'])
X.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import precision_score,recall_score
from sklearn.decomposition import PCA
from tempfile import mkdtemp
from shutil import rmtree
import eli5
from sklearn.svm import SVC,LinearSVC
from sklearn.metrics import make_scorer
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Logistic Regression

```python
cashdir=mkdtemp()
pipe_num=Pipeline([('scaler',StandardScaler()),("pca",PCA())])
col_transformer=ColumnTransformer([('encoder',OrdinalEncoder(categories=[X['gender'].unique(),X['A2'].unique(),X['A3'].unique()]
),['gender','A2','A3'])],remainder=pipe_num)
pipe_main=Pipeline([('transfomer',col_transformer),('classifier',LogisticRegression(class_weight='balanced'))]
                   ,memory=cashdir)
```

```python
pipe_main.get_params()
```

```python
param_grid={'classifier__C':np.logspace(-5,5,20),'transfomer__remainder__pca__n_components':range(2,20)}
grid_search=GridSearchCV(pipe_main,param_grid,cv=5,n_jobs=-1,scoring='precision',verbose=1)
grid_search.fit(X_train,y_train)
rmtree(cashdir)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
precision_score(y_test,grid_search.predict(X_test))
```

```python
recall_score(y_test,grid_search.predict(X_test))
```

```python
grid_search.best_estimator_.named_steps['classifier'].coef_
```

```python
grid_search.best_estimator_.named_steps.classifier.classes_
```

```python
eli5.show_weights(grid_search.best_estimator_.named_steps.classifier,feature_names=X_train.columns.to_list())
```

### Support Vector Classifier

```python
cashdir=mkdtemp()
pipe_num_svc=Pipeline([('scaler',StandardScaler())])
col_transformer_svc=ColumnTransformer([('encoder',OrdinalEncoder(categories=[X['gender'].unique(),X['A2'].unique(),X['A3'].unique()]
),['gender','A2','A3'])],remainder=pipe_num_svc)
pipe_main_svc=Pipeline([('transfomer',col_transformer_svc),('svc-classifier',SVC(kernel='linear',class_weight='balanced'
                                                                                 ,random_state=42))],memory=cashdir)
```
